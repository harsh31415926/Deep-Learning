```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab/sonar_dataset.csv", header = None)
df.head(4)
```

```python
df.shape
```

```python
X = df.drop(60 , axis = 'columns')
y = df[60]
```

```python
X.head()
```

```python
y = pd.get_dummies(y, drop_first = True)
y.replace({True : 1 , False : 0} , inplace = True)

y.head()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train , y_test = train_test_split(X , y , random_state = 42 , test_size = 0.2)
```

```python
X_train.shape
# X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60 , input_dim = 60 , activation = 'relu'),
    keras.layers.Dense(30 , activation='relu') ,
    keras.layers.Dense(15 , activation='relu') ,
    keras.layers.Dense(1 , activation='sigmoid')
])

model.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])

model.fit(X_train , y_train , epochs = 100 , batch_size = 8)
```

```python
yp = model.predict(X_test)

model.evaluate(X_test , y_test)
```

```python
y_pred  =  []

for i in yp:
  if i > 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)

y_pred = np.array(y_pred)
y_pred , y_test
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test , y_pred))
```

# With DropOut Layer

Now we are going to run the same model but with Dropout Layer
We got F1 score of 0.9

```python
modeld = keras.Sequential([
    keras.layers.Dense(60 , input_dim = 60 , activation = 'relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(30 , activation='relu') ,
    keras.layers.Dropout(0.3),
    keras.layers.Dense(15 , activation='relu') ,
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1 , activation='sigmoid')
])

modeld.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])

modeld.fit(X_train , y_train , epochs = 100 , batch_size = 8)
```

```python
yp = modeld.predict(X_test)

model.evaluate(X_test , y_test)
```

```python
print(classification_report(y_test , yd))
```
