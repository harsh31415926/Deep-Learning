```python
def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```python
def prediction_function(age , afford):
  weighted_sum = coef[0]*age + coef[1]*afford + intercept
  return sigmoid(weighted_sum)

prediction_function(0.58,1)
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))

def sigmoid_numpy(x):
    return 1/(1+ np.exp(-1))

sigmoid_numpy([1,2,4,2])
```

```python
def gradient_descent(age , afford , epoch , y_true , loss_thresold):
  w1 , w2 = 1 , 1
  bias  = 0
  rate = 0.02
  n = len(age)

  for i in range(epoch):
    weighted_sum = w1*age + w2*afford + bias
    y_predicted = (sigmoid_numpy(weighted_sum))

    loss = log_loss(y_true , y_predicted)
    w1d = (1/n)*np.dot(np.transpose(age),(y_predicted-y_true))
    w2d = (1/n)*np.dot(np.transpose(afford),(y_predicted-y_true))

    bias_d = np.mean(y_predicted-y_true)
    w1 = w1 - rate * w1d
    w2 = w2 - rate * w2d
    bias = bias - rate * bias_d

    print (f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')

    if loss<=loss_thresold:
      break

  return w1, w2, bias
```

# My own NN from scratch

Here we are going to create a working neural network without using tensorflow or keras , we will only use pure python for this

```python
class myNN():
  def __init__(self):
    self.w1 = 0
    self.w2 = 0
    self.bias = 1

  def fit(self , x,y , epoch , loss_thresold):
    self.w1 , self.w2 , self.bias = gradient_descent(x['age'] , x['affordibility'] , epoch ,y, loss_thresold)
    print(f'The value of W1 : {self.w1} and W2 : {self.w2} and bias: {self.bias}')

  def predict(self ,x_test):
    weighted_sum = self.w1*x_test['age'] + self.w2*x_test['affordibility'] + self.bias
    return sigmoid_numpy(weighted_sum)
```
